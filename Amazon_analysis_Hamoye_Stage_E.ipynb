```python
year = df1.groupby('year')['title'].count().nlargest(10).reset_index()
year
```

```python
title = df1.groupby('title')['no_of_reviews'].sum().nlargest(10).reset_index()
title
```

```python
fig = plt.figure(figsize = (15,8))
plt.bar(title['title'], title['no_of_reviews'])
plt.xticks(rotation = 45, ha = 'right')
plt.title('Top 10 books with the most reviews', size = 20)
plt.xlabel('Books', size = 12)
plt.ylabel('Number of reviews', size = 12)
plt.show()
```

This shows that the books 'Where the crawdads sing' and 'The girl on the train' are the books with the most reviews.  

```python
rating = df1.groupby(['title'])['ratings'].sum().nlargest(10).reset_index()
rating
```

```python
fig = plt.figure(figsize = (15,8))
plt.bar(rating['title'], rating['ratings'])
plt.xticks(rotation = 45, ha = 'right')
plt.title('Top 10 books with the highet ratings', size = 20)
plt.xlabel('Title', size = 12)
plt.ylabel('Ratings', size = 12)
plt.show()
```

This shows that 'StrengthsFinder 2.0' and 'The official SAT study Guide' are the books with the highest rating.

```python
rating_s = df1.groupby(['title'])['ratings'].sum().nsmallest(10).reset_index()
rating_s
```

```python
fig = plt.figure(figsize = (15,8))
plt.bar(rating_s['title'], rating_s['ratings'])
plt.xticks(rotation = 45, ha = 'right')
plt.title('Least 10 books with the least ratings', size = 20)
plt.xlabel('Title', size = 12)
plt.ylabel('Ratings', size = 12)
plt.show()
```

Here, it can be seen that 'The casual vacancy'and 'Go set a watchman: A Novel' are the books with least ratings.
